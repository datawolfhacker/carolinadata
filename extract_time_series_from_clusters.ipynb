```python
import numpy as np
import pandas as pd
```

```python
# read the csv file
df = pd.read_csv("preprocessed_NYPD_Complaint_Data_Historic_short.csv")
```

```python
df.columns = map(lambda x: x.lower(), df.columns)
```

```python
df[:15]
```

```python
region_info = pd.read_csv('precinct_mean.csv')
```

```python
region_info[:40]
```

```python
df['cluster_label'] = 0
```

```python
for i in df['addr_pct_cd'].unique():
    df['cluster_label'][df['addr_pct_cd']==i] = int(region_info['cluster_label'][region_info['addr_pct_cd']==i])
```

```python
df[:20]
```

```python
aggregated_by_date_df = df.groupby(['cluster_label', 'cmplnt_fr_dt']).count()['addr_pct_cd']
```

```python
for i in range(10):
    (aggregated_by_date_df[i]).to_csv('time_series/time_series_of_cluster_{}.csv'.format(i))
```

```python
for i in range(10):
    df = pd.read_csv('time_series/time_series_of_cluster_{}.csv'.format(i))
    df.columns = ['date', 'count']
    df['date'] = df['date'].apply(lambda x: x[:-4] + str(int(x[-4:]) + 1000) if int(x[-4:]) < 1900 else x)
    df.to_csv('time_series/time_series_of_cluster_{}.csv'.format(i), index=False)
```
